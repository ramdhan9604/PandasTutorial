```python
# Q. find ranking of each movie in the group according to IMDB rating
```

```python
def movie_rank(group): 
    group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(movie_rank)
```

```python
# Q. find normalized IMDB rating group_wise
```

```python
def normal(group): 
    group['norm_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal)
```

```python
ipl = pd.read_csv('deliveries.csv')
```

```python
ipl.shape
```

```python
ipl.head()
```

```python
# Q.1. Find the top 10 batsman in terms of runs
```

```python
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Q.2. find the batsman with max no of sixes
```

```python
six = ipl[ipl['batsman_runs']==6]
```

```python
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Q.3. find batsman with most number of 4's and 6's in last 5 overs
```

```python
temp = ipl[ipl['over']>15]
```

```python
temp_runs = temp[(temp['batsman_runs']==4 ) | (temp['batsman_runs']==6)]
```

```python
temp_runs.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Q.4. Find V Kohli runs against all teams
```

```python
temp = ipl[ipl['batsman']=='V Kohli']
```

```python
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Q.5. create a function that can return the highest score of any batsman
```

```python
temp = ipl[ipl['batsman']=='V Kohli']
```

```python
temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
    temp=ipl[ipl['batsman']==batsman]
    return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('MS Dhoni')
```

```python
highest('RG Sharma')
```
